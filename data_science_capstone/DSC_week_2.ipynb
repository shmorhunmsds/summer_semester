## 2.1 Lesson: Lasso, Ridge, and Elastic Net Regression

### Lasso Regression:
Suppose you are trying to match the prices of 100 common grocery items to user expenditures. You dont want to do a linear regression with 100 features  you're concerned that youll overfit, and you also want to report something meaningful to your employer. Out of your 100 items, you want to drop many or most of them and only consider the important ones. Lasso regression is a good choice for this task. 

Lasso regression is a regularized version of regression. Instead of minimizing the error of a prediction relative to a true target value, it also tries to minimize a regularization term.
- The regularization term is equal to a constant $\lambda$ (lambda) multiplied by the sum of the absolute values of the coefficients.
- The goal is to make the coefficients as small as possible.
- For example:
    - The expression $Y = X_1 + X_2$ and $Y = 10X_1 - 8X_2$ both explain the data,
    - In lasso regression the former equation ($Y = X_1 + X_2$) is better because the coefficients are smaller. 
- In fact, even if the $Y = X_1 + X_2$ is not as good at predicting the target value $Y$, it might still be selected by the lasso regression because the coefficients are smaller.

Lasso tends to make some coefficients go to zero. For example:

Suppose the regularization term is $\lambda (\beta_1 + \beta_2 + \beta_3)$

Then, if we reduce $\beta_1$ by one point, how much does the fit worsen?

If it worsens by lambda points or less, then we should perhaps keep reducing $\beta_1$ until it reaches zero. That's beacuse the fit gets better by lambda points (due to the regularization term) and then worse by less than lambda points for a net benefit. In that case, maybe its not worth it to have a $\beta_1$ after all.

Lasso handles multicollinearity by, in essence, dropping one of the features at random. Many people feel that this is too arbitrary, in which case, you might want ridge regression.

### Ridge Regression:
Going back to our grocery store scenario, now suppose that you do want to include all 100 grocery store items. You are okay with including as many as possible, but you still want to reduce the impact of the items that are unimportant. Youd ideally end up with a list of 100 coefficients, of which a few are large, but most are small. Ridge regression is an ideal approach for you in this case. 

Ridge regression is like lasso regression, but instead of using the sum of the absolute values, it uses the sum of the squares of the coefficients. Ridge regression tends to keep many nonzero coefficients.

> Suppose we have a new regularization term that is $\lambda * (\beta^2_1 + \beta^2_2 + \beta^2_3)$. 

> If we reduce $\beta_2$ by one point, we reduce the regularization term by $2 \cdot \lambda \cdot \beta_1$ points.

> But as $\beta_1$ shrinks, this improvement becomes more and more insignificant, so eventually it's woth keeping the $\beta_1$ term, so long as $\beta_1$ is small enough.

Ridge regression usually handles multicollinearity by assigning a nonzero term to each feature.

#### Selecting an Optimal Lambda Parameter
In applying both ridge and lasso regression, the way to select the lambda parameter is to try different values, using some sort of hyperparameter search. For example, you could use cross-validation with a grid search or random search. It might be best to try lambda values in a logarithmic space, i.e., trying $\lambda$ = 100, 10, 1, 0.1, 0.01, etc. 

### Elastic Net Regression
Now consider the earlier grocery scenario:
- You want to compromise between the lasso and ridge approaches. You've tried both, but lasso cust out too many parameters, while ridge leaves too many of them in. In this situation, elastic net is an option. 
- Elastic net regression uses an $\alpha$ (alpha) parameter to interpolate betwen lasso and ridge. Y ou can be partway between lasso and ridge. This could allow you to perform feature selection (like lasso) but also handle multicollinearity in the way that ridge does.

### Summary:
- Lasso regression is a regularized version of regression. Instead of minimizing the error of a prediction relative to a true target value, it also tries to minimize a regularization term. 
- Ridge regression is like lasso regression, but instead of using the sum of the absolute values, it uses the sum of the squares of the coefficients. Ridge regression tends to keep many positive coefficients. 
- Elastic net regression uses an alpha parameter to interpolate between lasso and ridge.
