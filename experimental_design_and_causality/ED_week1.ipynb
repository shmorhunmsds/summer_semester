## 1.1 Lesson: Causal Inference
This course is about causal inference — the science of using data to understand what causes what. In the following videos, you'll explore how data and domain knowledge can be combined to estimate causal effects — not just correlations. You'll learn how confounding variables can obscure the true relationship between cause and effect, why domain expertise is critical to identifying those confounders, and how methods like regression and matching help us untangle these relationships. As you watch, consider how your own knowledge of a domain can help identify what data are needed—and what assumptions must be made—to draw meaningful causal conclusions. 

As you learned in the videos, a classic example of causal inference, or the study of relationships between variables, is the relationship between shark attacks and ice cream sales. It’s a true fact that when ice cream sales are high, shark attacks are common as well. This may suggest the idea of outlawing ice cream in order to protect swimmers. Is that a good idea? Probably not. First of all, common sense tells us that there is no way that ice cream consumption would cause shark attacks. If someone tells us that there is, we should be skeptical. Second, it may be clear why this correlation exists: ice cream sales are high in the summer — and people go swimming in the summer. In the winter, fewer people eat ice cream, and no one goes swimming at the beach. So, we’d expect ice cream sales to be high precisely when shark attacks are common — that is, in the summer. The two events have a common cause, that is, the heat of the summer. 

What would happen if we reduced ice cream consumption anyway? Would shark attacks decrease? No — the shark attacks are caused only by the summer heat. Reducing ice cream consumption would not reduce the summer heat, and so shark attacks would be unaffected. 

Let’s consider another example. We have a medicine that is meant to treat high blood pressure. We find 100 patients who are taking the medication and 100 who are not. Of these, 70% of those taking the medication have high blood pressure, but only 40% of those not taking the medication have high blood pressure. Does this mean that the medication causes high blood pressure? No! Surely the explanation is that people with high blood pressure are more likely to take the medication. 

If we took a random person and gave them the medication, it is possible (though we cannot yet prove it) that it would actually reduce their blood pressure. We don’t know yet. How can we find out? There are two approaches we can take: regression and matching. Suppose we want to isolate the causal effect of the medication. How can we do that? Here are two ways. 

### The Experiment
We could run an experiment:
- Take 200 people. 
- Randomly assign 100 of them to the “treatment” group (they’ll get the medication) and 100 to the “control” group (they won’t get it). 

Now, the two groups are probably roughly the same in every way: 
- They have approximately the same blood pressure, the same age, the same height and weight, and so on. 
- They are, in some sense, “matched” in that for every tall person in the treatment group, there is (probably) a tall person in the control group. 
- It’s probabilistic, of course; maybe the treatment group’s average height is 5 feet 6 inches, while the control group is 5 feet 7 inches or even 5 feet 8 inches. 
- Maybe one of the groups, by chance, has a very tall person in it who throws off the average. But if we have enough participants, then everything will average out, the participants will be a true representation of the overall population, and the groups will be (in effect) matched with one another. 
- In that case, we can look at the average blood pressure of the treatment group and the average blood pressure of the control group after taking the medication. If they are different (and statistically significantly different — we need the big sample size to get statistical significance), then we can confidently say that the medication probably helped. 

### What If No Sample Size Is Big Enough? 
As an aside: It is possible, in principle, for this kind of experiment to fail, even if we have a very large sample (and it could fail no matter how big we make the sample). However, this possibility is a weird mathematical abnormality that never happens in real life. Let’s consider the possibility anyway. 

This kind of weird situation would involve the “Pareto distribution” or a similar distribution.

![Relationship between Sample size & Maxtree.png](<attachment:Relationship between Sample size & Maxtree.png>)

Suppose we want to estimate the mean tree height for a certain species of tree. 
- Suppose that most trees are 10 feet tall
- approximately 1 in every 100 trees is over 20 feet tall, 
- 1 in every 200 trees is over 40 feet tall, 
- 1 in every 400 trees is over 80 feet tall, 
- 1 in every 800 trees is over 160 feet tall, and so on. 

Each time we multiply the sample by two, the typical maximum height doubles too. This is an example of a **Pareto probability distribution**.

Then, in 100 trees, a typical mean height might be (with 99 regular trees and one tall one):

$ 0.99 \times 10 + 0.01 \times 20 \; = \; 10.2 \; \text{feet} $

In 200 trees, a typical mean height might be (with 198 regular trees, one tall one, and one really tall one): 

$0.99 \times 10 + 0.005 \times 20 \times 40 \; = \; 10.2 \; \text{feet} $

In 400 trees, a typical mean height might be (with 396 regular trees, two tall ones, one really tall one, and one really, *really* tall one): 

$0.99 \times 10 + 0.005 \times 20 + 0.0025 \times 80 \; = \; 10.3  \;\text{feet}$

As you can see, the bigger we make the sample, the taller the typical mean tree height gets. Now, surely this must stop eventually! Is it really true that 1in every 100,000 trees is at least 20,480 feet tall? But in principle, if it didn’t stop, the typical mean tree height would grow bigger as you add more trees.  

What do I mean by “typical mean height”? I mean the median across all possible samples. In other words, suppose you took 1,000 samples of 100 trees each. Some of them would have a mean of 9 feet, some 10.5 feet, some 11.3 feet. 

The median of these means would be something like 10.1 feet and would stay fixed no matter how many samples of 100 trees we took. 

The mean of the means would increase without limit as we take more and more samples of 100 trees. Weird, huh? 

As long as there aren’t successive abnormal cases that get exponentially more extreme as we take larger and larger samples, we can eventually get a large enough sample that we’ve covered all possible cases. 

Of course, in real life, we may be limited as to how big a sample we can take. 

So, the fact that we could, in principle, get a big enough sample doesn’t necessarily help; we live in the real world, where our sample size is limited.

## Chapter 1 - Designing Research

### 1.1 I have a question
A **research question** is a question that you have that you plan to answer, or at least try to answer, by doing research. Simple as that. Or, rather, as difficult as that. A good research question is well-defined, answerable, and understandable

- let’s say that our research question is “does adding an additional highway lane reduce traffic?”

### 1.2 Empirical Research
This book will focus on empirical research and, specifically, **quantitative empirical research**. Empirical research is any research that uses structured observations from the real world to attempt to answer questions. 
- So instead of trying to *reason* our way through what drivers would do if given an additional highway lane, we try to *observe* the choices that drivers take. Perhaps we interview drivers about how they make decisions. Or maybe we get a big dataset of traffic violations, or of traffic flow numbers on highways.

One particularly sticky problem with quantitative empirical research is that the numbers that we observe often don’t tell us exactly what we want to know.

*We’re probably interested in whether we can make traffic go down by turning a two-lane highway into a three-lane highway! But as much as we want them to, the numbers we have don’t actually tell us that right away. All we have are two-lane highways and three-lane highways. We don’t have a “what if” highway that tells us how much traffic there would have been if we’d made that two-lane highway one lane wider.*

This problem constitutes a major headache for us researchers. If the numbers we have don’t actually answer the research question we have, what can we do?

Well, it turns out that, if you do it right, you often can figure out how to collect the right numbers, or do the right things to those numbers, to get an actual answer to our question. But it doesn’t come free. We have to carefully design the right kind of analysis that will answer our question.

### 1.3 Why Research Needs a Design
Why is it so important for research to be properly designed? One way we can think about this is by looking at what happens when it’s not.

Let’s take our highways and traffic example. How might we go about researching an answer to this question?

*Our first pass might be to just compare traffic patterns on highways with more lanes against traffic patterns on highways with fewer lanes. Seems reasonable. But then you do it, and it turns out that more lanes seem to go along with more traffic!* 

*Sure, maybe additional lanes do lead to more traffic. But it takes research design to know that our first-pass analysis wasn’t right and to figure out what to do instead.*

A lack of solid research design can be seen in the results, as well.

## Chapter 2 - Research Questions

### 2.1 What Is a Research Question?
Coming up with a question is easy. Coming up with a good research question is much harder.

**A research question is a question that can be answered, and for which having that answer will improve your understanding of how the world works.**

What does it mean to have a question that can be answered? It means that it’s possible for there to be some set of evidence in the world that, if you found that evidence, your question would have a believable answer.

So we have a question that can be answered. But does it improve our understanding of how the world works? 

What this means is that the research question, once answered, should tell you about something broader than itself. It should inform theory in some way.

Theory just means that there’s a why or a because lurking around somewhere.

Take germ theory, for example. Germ theory says that microorganisms like bacteria and viruses can cause disease. This explains why we have diseases, and also why disease can spread from one person to another. We don’t call it “a theory” because we're uncertain about whether its true. We call it a theory because it tells us *why*.

A good research question takes us from theory to hypothesis, where a hypothesis is a specific statement about what we will observe in the world. That is, a research question shoudl be something that, if you answer it, helps improve your *why* explanation. Great research questions often come from the theory themselves. 

Let’s check our two conditions for research questions.
1. Could we answer this question?
2. Second, does this research question tell us about how the world works?

A good test for whether a research question informs theory is to imagine that you find an unexpected result, and then wonder whether it would make you change your understanding of the world.

This ability to see a bad result and still hold on to the original theory tells us that the research question wasn’t very good, at least not for this theory.66 This means if the Sesame Street study had turned out in favor of our theory we shouldn’t have increased our confidence either. 

A really good research question, once answered, should be hard to explain away just because it’s inconvenient.

### 2.2 Why Start with a Question?
why not skip the hard part of deriving a research question from a theory and instead just see what sorts of patterns are in the data? This is called *data mining*, and it turns out to be very good at somethings and very bad at others.
- Data mining is good at *finding patterns* and in *making predictions under stability*
- Data mining is less good at *improving our understanding*, (or helping to improve theory)
- Data mining also has a tendency to find *false positives* if you aren't careful.

Finding patterns and making predictions are very valuable. After all, there’s no way we can really theorize about every possible pattern that could be in the data and think to check it. Doing something that just asks what we see rather than why is the right angle to take there. Plus, sometimes seeing patterns in data can give us ideas for research questions that we can examine further in other data sources.

#### Why does Data Mining Have Difficulty Helping Theory?
- Data mining focuses on what's in the data, not *why* its in the data. 
- Its fantastic at revealing correlations - patterns in the data of how variables we've observed have varied together in the past - but the correlations it uncovers may have little to do with causality, or an understanding of why those variables move together. 
- Data mining is well-equipped to find the relationship but poorly-equipped to tell us why that relationship is there.
- Another reason is that, because it’s so focused on the data, data mining doesn’t really deal in abstraction.

#### False Positives
False positives are another reason why data mining can be dangerous. There are ways of avoiding false positives while doing data mining - this is something they worry about a lot in data science and have a lot of tools for. Data mining isn’t bad. It’s just bad as a final step if you’re trying to explain the world.

### Where do Research Questions Come From?
Research questions can come from lots of places. There are two steps in this process: 
1. thinking about theory
2. coming up with a research question. 

Either one can come first.

Perhaps it begins with theory. With the theory in place, the process continues with our hypothesis:

> *“if this is how the world works, what would I expect to see in the world?”*

Our theories might lead to research questions. These research questions tel us a hypothesis to test such that the result of that test tells us something about the theory. 


Let’s be honest, sometimes research questions also come from opportunity. Have a neat data set? Think about what data is available to you and whether any related research questions or theories come to mind.

### How do you know if you've got a good research question?
But is it really a good one? Just a few things to check before you get too far into the process:
1. Consider Potential Results. 
- A good way to double-check the relationship between your research question and your theory is to consider the potential answers you might get. 
- Then, imagine what kind of sense you’d make of that result, or what conclusion you would draw. Let’s say you find that students do tend to work harder in school when they’re paid for good grades. What would this tell us about how students respond to incentives? Or let’s say you find that students don’t work harder when they’re paid. What would that tell us about how students respond to incentives? 
- If you can’t say something interesting about your potential results, that probably means your research question and your theory aren’t as closely linked as you think! Let’s say we do find that kids who happen to play video games are more aggressive. Can we take that result and claim that video games are a cause of aggression? Not really, for the reasons we’ve discussed previously. So maybe that research question really isn’t linked to that theory very well.

2. Consider Feasibility
- A research question should be a question that can be answered using the right data, if the right data is available. But is the right data available? 
- If answering your research question is possible but requires following millions of people repeatedly for decades, or trying to measure something that’s really hard to measure accurately, like trying to get people to remember what they had for lunch three years ago, or getting access to the private finances of thousands of unwilling people, then that research question might not be feasible. 
- While sometimes you can get around these problems with a clever design, you might want to consider going back to the drawing board.

3. Consider Design
- A research question can be great on its own, but it can only be so interesting without an answer. So, an important part of evaluating whether you have a workable research question is figuring out if there’s a reasonable research design you can use to answer it. Figuring out whether you do have a reasonable research design is the topic of the rest of this book.

4. Keep it Simple
- Answering any research question can be difficult. Don’t make it even harder on yourself by biting off more than you can chew! 
- A common mistake is to bundle a bunch of research questions into one. “What are the determinants of social mobility?” I.e., how someone can move from one social class to another throughout their lifetime. There are many determinants of social mobility. You’re unlikely to answer that question well. Instead try “Is birth location a determinant of social mobility?” For another example, how about the question “How was the medium of painting affected by the Italian renaissance?” In a million ways! You’ll get lost and do a poor job on a bunch of minor pieces instead of getting at the whole. Instead maybe “What similar characteristics are there among the countries that adopted the use of perspective in painting most quickly?”

So, consider feasibility, scale, and design. Keep it simple, and think about whether the results you might likely see would tell you anything interesting about the world. After all, learning something interesting and new about the world is our goal!


## 1.2 Lesson - Coefficient Estimates

### What if We Didn't Do an Experiment?
But what if we didn’t do a controlled experiment where we take 200 people and randomly assign 100 of them to the “treatment” group? We have some data, but it doesn’t come from an experiment. It comes from something else, some non-random assignment of people into treatment and control groups. 
- Perhaps we just collected some hospital records and looked at the first 200 people in the records. 
- But the set of people who enter a hospital are non-random. They’re more likely to have a serious illness, for one thing. 
- They may be older, on average, than the general population, except that new babies are also overrepresented (since they are likely to be born in the hospital.) 

This is not necessarily a problem, though. We could still do a controlled experiment on all the people in the hospital, and then we would find out the medication’s effectiveness on hospital patients. There’s nothing wrong with that! Maybe it is interesting to know the effectiveness of the medication on hospital patients.

But there’s a deeper problem: the medication wasn’t given to people at random. In fact, it was likely given to people with high blood pressure. 

Then the blood pressure is a **“confounder,”** meaning that it influenced two things: 
1.  the outcome (the initial blood pressure and final blood pressure are related), and 
2.  whether the medication was given at all (a high initial blood pressure was more likely to lead to being treated.) 

Because this confounder influences both the treatment and the outcome, there will be two separate reasons for the outcome to relate to the treatment:
1. Maybe the treatment causes the outcome (the medication reduces your blood pressure). 
2. Maybe the treated participants are more likely to have a high initial blood pressure (before taking the medication), and, therefore, they have a higher final blood pressure (after taking the medication).
___

This is a serious problem for our study. How will we know whether our data reflect results (1), (2), or a mixture of both? It seems likely that it’s a mixture of both — but what mixture, exactly? Once again, there are two ways to address the problem: regression and matching.

The first approach is to assume that there is a *linear relationship* between the outcome (blood pressure after medication) and the two other factors (whether the person is treated or not and whether the person had high blood pressure before being treated). 

___
For example, suppose the medication actually subtracts 10 mm Hg from someone’s systolic blood pressure. Then, if $X$ is whether the person is treated (0 for untreated, 1 for treated), $Z$ is the person’s initial blood pressure, and $Y$ is the outcome (their final blood pressure), the formula is:

$Y \; = \; Z - X \times 10$

That’s a linear formula. Of course, it’s probably not quite so deterministic. 

For example, biological formulas might vary day to day depending on what the person eats or how much sleep they get. 

There’s also some random variable, let’s call it epsilon. 

Then:

$Y \; = \; Z - X \times 10 + \epsilon$

For instance, suppose the initial blood pressure ($Z$) is $115$, and the person is treated ($X \; = \; 1$), but the final blood pressure ($Y$) is $107$. 

Then:

$Y(107) \; = \; Z(115) - X(1) \times 10 + \epsilon(2)$

In other words, the "error" $\epsilon$ was $2$, so $107 \; = \; 15 - 1 \times 10 + 2$. To get rid of the error (or the amount that remains after calculating the linear formula in $X$ and $Z$), we'd have to do the experiment a bunch of times and take some averages. 

So our linear formula has a coefficient of $-10$ (for $X$), meaning that it contains the term $-10X$.

It also has a coefficient of $1$ (for $Z$), meaning that it contains the term $Z$, or $1 * Z$.

Our goal in linear regression is to deduce these coefficients. Our formula does not contain an intercept; that is, there is no additional constant that is added to the equation.

This equation shows that if you treat someone, their blood pressure should decrease by 10. We do not have the causal relationship that we want. 

Of course, in our data, if we take the treated and untreated groups, their mean blood pressure will not differ by 10. It will likely differ by less (or the treated group could even have a higher final blood pressure) because the treated group has a higher value of $Z$. That’s why we needed to do a linear regression in order to sort out these different factors.

The coefficient of $X$ (which is $10$) is the *effect* that's mentioned in the module's book title "The effect". It tells us *what happens when you treat a particular individual*. 

Of course, at least one problem remains: What if there are other confounders besides the initial blood pressure ﻿Z﻿? What if there’s some ﻿W﻿ (like weight, recent exercise, or age) in the equation, too? Another problem is: What if there’s a nonlinear relationship of some kind that can’t be written out in this linear fashion? We’ll discuss these issues in a later week.

Through this approach, we aim to increase the accuracy of our findings and reduce potential bias in experimental results.

## 2.1 Lesson: Matching with 1:1 Correspondence
Let’s revisit the experiment we discussed earlier this week, where we compared the effects of a new drug on patients by using a treatment group and a control group. 

Now, let’s dive deeper into another approach we could take: For each person in the treatment group (those who take the medication), pick one person in the control group who has about the same $Z$ value (the same initial blood pressure). Now, compare the treatment group’s outcome to this “matched” control group. At that point, we’re doing an apples-to-apples comparison and comparing two groups who have the same blood pressure. That’s exactly what we’d get if we did a true controlled experiment. 

For example, if the two groups’ $Z$ values (initial blood pressures) are:

`treatment = [120, 125, 130, 135, 140]`

`control = [100, 110, 120, 130, 140]`

Then, we could match each treatment group’s $Z$ values to one in the control:

| Treatment Z | Control Z |
| :--- | :--- |
| 120 | 120 |
| 125 | 130 |
| 130 | 130 |
| 135 | 140 |
| 140 | 140 |

This lets us create two groups of five samples each (although the control group now has duplicates), since the `control Z = 130`  and `control Z = 140` were used twice.

The distribution of $Z$ values in each group is about the same, so the effect we measure is going to be meaningful.

The effect we measure is connected to our *distribution* of $Z$ values. If our treatment group had a different distribution of $Z$ values, there might be a different effect.

That’s always going to be a problem, no matter what! As we said, the effect in a hospital might not be the same as the effect in an academic study because the two groups are different. There’s no way to escape that: Different groups are different.

But at least now we know the effect in one type of group, the group with $Z$ values of about 120, 125, 130, 135, and 140.

We could also take subsamples of this group and see how they behave (say, taking just the $Z > 130$ values or the $Z < 130$ values).

There is, once again, a problem that there may be some other confounder besides $Z$. We just have to make our best argument about whether that is or isn’t the case and then hope we are right. There is likely no way to be absolutely sure that $Z$ is the only confounder, but we can do our best to figure it out.

Also, we have to decide exactly how to do the matching. Will we pick the one best control item for each treatment item? Is it possible that, if there is no very good control item to match, we might not be able to include a particular treatment item after all? What if someone in one group has a very high or very low blood pressure, and there is no matching person in the other group? We will discuss these issues in a later week.
